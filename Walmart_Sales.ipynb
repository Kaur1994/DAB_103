# RETAIL SALES ANALYSIS  USING WALMART DATASET BY TEAM DATACREW
 DATASET : https://www.kaggle.com/rutuspatel/retail-analysis-with-walmart-sales-data


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wal_d = pd.read_csv('/content/WALMART_SALES_DATA (1).csv')
wal_d.head()
```

```python
wal_d.info()
```

```python
wal_d['Date'] = pd.to_datetime(wal_d['Date'])
wal_d['year'] = pd.DatetimeIndex(wal_d['Date']).year
wal_d['month'] = pd.DatetimeIndex(wal_d['Date']).month
```

```python
wal_d.head()
```

```python
wal_d.tail()
```

# EXPLORATORY DATA ANALYSIS OF WALMART SALES DATASET VARIABLES

```python
plt.figure(figsize=(6, 6))
sns.barplot(x=wal_d['Holiday_Flag'],y=wal_d['Weekly_Sales'],color='#8B1C62')
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.lineplot(data=wal_d,x='month',y='Fuel_Price',hue='year',style='year',palette='dark')
plt.title("FUEL PRICE PATTERN BY MONTH FOR 3 CONSECUTIVE YEARS - 2010 , 2011 , 2012")
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.lineplot(data=wal_d,x='month',y='Unemployment',hue='year',style='year',markers=True,palette='colorblind')
plt.title("UNEMPLOYMENT RATE BY MONTH OF SALES RECORDED YEARS -2010 , 2011 , 2012")
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.lineplot(data=wal_d,x='month',y='Temperature',hue='year',style='year',markers=True,palette='bright')
plt.title("TEMPERATURE VARIATION BY MONTH OF SALES RECORDED YEARS -2010 , 2011 , 2012")
plt.show()
```

```python
sns.pairplot(wal_d[wal_d['year'] == 2012], 
             vars = ['Weekly_Sales', 'Temperature', 'Unemployment'], 
             hue = 'month',
             size = 7,palette='Paired')
```

```python
corr = wal_d.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr, annot=True)
plt.plot()
```
